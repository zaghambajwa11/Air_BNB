```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

df = pd.read_csv("/kaggle/input/ab-nyc/AB_NYC_2019.csv")
df.head()




```

 **Data cleaning and preprocessing: Before analyzing the data**
* it's important to clean and preprocess it to ensure that 
* it is accurate and useful for analysis. 
* Data cleaning involves identifying and fixing errors, 
* inconsistencies, and missing values in the dataset. 
* Preprocessing involves transforming the data into a format that is more suitable for analysis.

```python
# Removing the duplicates Values by using pandas function

df = df.drop_duplicates()
```

```python
# First Check how many values are missing
# count the number of missing values
num_missing = df.isnull().sum().sum()

print("Number of missing values:", num_missing)

# Number of missing values: 20141.. when I Ran above line. these are total vNull Values 
```

```python
# Now check which rows contain null values
rows_with_null = df.isnull().any(axis=1)

# count the number of rows with null values
num_rows_with_null = rows_with_null.sum()

print("Number of rows with null values:", num_rows_with_null)

#Number of rows with null values: 10074
```

```python
#Lets Calculate the total rows in the datafram
# get the total number of rows in the dataframe
num_rows = df.shape[0]

print("Number of rows in the dataframe:", num_rows)
```

```python
#Lets creat stacked bar chart

total_value = 48895
value_1 = 20141
value_2 = 10074

values = [total_value,value_1 ,value_2]
labels = ['Total Rows', 'Total Null Values', 'Total Rows With Null Value']


# add values on top of bars
for i, v in enumerate(values):
    plt.text(i, v + 1, str(v), ha='center')

# create a vertical bar char
x = np.arange(len(labels))
plt.bar(x, values)



# add chart labels
plt.ylabel('Value')
plt.xticks(x, labels)


# show the chart
plt.show()
```
